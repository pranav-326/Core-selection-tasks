```python
pfrom transformers import AutoModelForCausalLM, AutoTokenizer

from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype="auto"
)

model.config.use_cache = False
```

```python
from datasets import load_dataset

dataset = load_dataset(
    "json",
    data_files="/content/drive/My Drive/ieeeProject/event_reports.jsonl"
)["train"]
print(len(dataset))
print(dataset[0])
```

```python
from peft import LoraConfig, get_peft_model

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    args=SFTConfig(
        dataset_text_field="text",
        per_device_train_batch_size=2,
        gradient_accumulation_steps=2,
        num_train_epochs=5,
        learning_rate=2e-4,
        logging_steps=5,
        output_dir="./out",
        report_to="none"
    ),
)

trainer.train()
```

```python
prompt = """
Write an IEEE student chapter event report using the following details:

Event Name: AI Horizons 2026
Organizing Body: NIE IEEE Student Branch
Date: 12 January 2026
Event Type: Technical Talk
Venue: NIE Campus
Objective: To introduce students to emerging AI trends.
Activities: Expert lecture on AI trends; Q&A session; Case study discussion.
Outcome: Provided clarity on AI advancements and industry expectations.

Report:
"""

import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

inputs = tokenizer(prompt, return_tensors="pt").to(device)
model.to(device)

model.eval()
output = model.generate(
    **inputs,
    max_new_tokens=300,
    temperature=0.3,
    top_p=0.9
)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
merged_model = model.merge_and_unload()
merged_model.save_pretrained("ieee-report-generator-merged")
tokenizer.save_pretrained("ieee-report-generator-merged")

merged_model.push_to_hub("pranav-326/ieee-report-generator-merged")
```
